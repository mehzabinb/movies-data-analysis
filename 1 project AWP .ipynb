```python
mstar_rating=df['star_rating'].mean()
mstar_rating=df['star_rating'].median()

sns.distplot(df['star_rating'])
plt.axvline(mstar_rating,color='r')
plt.axvline(mstar_rating,color='y')
plt.grid(True)
plt.show()
```

```python
#Fetching order priority
```

```python
sns.countplot(df['star_rating']) 
# 7.6 has highest count followed by 7.8>7.7>7.5
# 7.7 and 7.5 have very minor difference and both are above 100
# 8.2 and 7.4 have very minor difference and both are above 40
```

```python
rates=df.groupby('star_rating').sum()['duration'].reset_index()
rates.sort_values('duration',ascending=False)
#All the items arranged according to their duration.
# 14548 is the highest duration and there star_rating is 7.6
# 142 is the lowest duration but their star rating is higest 9.3
```

```python
plt.figure(figsize=(10,5))
plt.plot(rates['star_rating'],rates['duration'])
plt.xticks(rotation='vertical',size=8)
plt.show()
```
